### Here, we plot the distributions of the scores before calibration and after

```python
import matplotlib.pyplot as plt
import os
import pandas as pd
import random

from functools import partial
```

```python
splits = ['train', 'validation', 'test']
path = "/home/li.mil/open-domain-mds-merge/open-mds/analysis/csv"
```

```python
def normalize(x, df, col):
    return (x - min(df[col])) / (max(df[col]) - min(df[col]))

full_path = os.path.join(path, "train_probs.csv")
df = pd.read_csv(full_path)

# Get the qid list
qids = list(set(df.qid))

# Select a random number of qids and plot their distributions -- let's say 10?
for i in range(10):
    qid = random.choice(qids)
    retrieved = df[df.qid == qid]
    result = pd.DataFrame()
    scores = retrieved['score']
    calibrated = retrieved['calibrated']
#         result['norm_score'] = retrieved['score'].apply(partial(normalize, df=retrieved, col="score"))
#         result['norm_calibrated'] = retrieved['calibrated'].apply(partial(normalize, df=retrieved, col="calibrated"))

    fig, ax = plt.subplots(1, 2, figsize=(9, 5))
#         ax[0].hist(result['norm_score'], alpha=0.5, color="red", label="pre-calibrated scores")
#         ax[1].hist(result['norm_calibrated'], alpha=0.5, color="green", label="calibrated scores")
    ax[0].hist(scores, alpha=0.5, color="red", label="pre-calibrated scores")
    ax[1].hist(calibrated, alpha=0.5, color="blue", label="calibrated scores")
    ax[0].set_title("Pre-calibrated scores")
    ax[1].set_title("Calibrated scores")
#         plt.hist(result['norm_score'], alpha=0.5, color="red")
#         plt.hist(result['norm_calibrated'], alpha=0.5, color="green")
    plt.show()
#     for qid in qids:
        
```
